# Decorators in Python
Decorators are very powerful and useful tool in Python since it allows programmers to modify the behavior of function or class.

## Uses Of Decorators
#--> To validate args
#--> To alter args to function
#--> Decides to execute or not execute function according to args.
#--> To call initialization & Downtype functions before & after function execution.
#--> To alter return type.



```python
def add_deco(func):
    print(func)
    def inner(a, b,c):
        print("I am going to add", a, "and", b, "and",c)
        a,b,c=10,20,45
        if b == 0:
            print("no operation")
            return
        # c = a*b    
        return func(a,b,c)
    return inner


@add_deco
def add(a, b,c):
    return (a,b,c,a+b+c)

print(add(40,4,8))
```
