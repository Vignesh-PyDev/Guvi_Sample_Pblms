# Matrix Addition & Subraction
Two matrices may be added or subtracted only if they have the same dimension; that is, they must have the same number of rows and columns.

```python
def Matrix_Dimension_Check(Mtrx1,Mtrx2):
    if(len(Mtrx1) == len(Mtrx2)):
        for x in range(len(Mtrx1)):
            if(len(Mtrx1[x]) == len(Mtrx2[x])):
                return 1
                break
            else:
                break
                return 0
                
```

```python
def Matrix_Add(Mtrx_1,Mtrx_2):
    
    Result_Matrix_Add = []
    if(Matrix_Dimension_Check(Mtrx_1,Mtrx_2)):
        for x in range(len(Mtrx_1)):
            for y in range(len(Mtrx_1)):
                Result_Matrix_Add.append((Mtrx_1[x][y] + Mtrx_2[x][y]))
        return Result_Matrix_Add
    else:
        print("Matrtix Size Mismatch!!!\nCannot Perform Desired Operation")
```

```python
def Matrix_Sub(Mtrx_1,Mtrx_2):
    
    Result_Matrix_Sub = []
    if(Matrix_Dimension_Check(Mtrx_1,Mtrx_2)):
        for x in range(len(Mtrx_1)):
            for y in range(len(Mtrx_1)):
                Result_Matrix_Sub.append((Mtrx_1[x][y] - Mtrx_2[x][y]))
        return Result_Matrix_Sub
    else:
        print("Matrtix Size Mismatch!!!\nCannot Perform Desired Operation")
```
